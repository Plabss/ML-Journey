```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Set parameters
mean = 0
std_dev = 1
num_samples = 10000

# 2. Generate data points from a normal distribution
data = np.random.normal(mean, std_dev, num_samples)

# 3. Plotting
plt.figure(figsize=(10, 6))
sns.histplot(data, kde=True, color='blue', stat='density')

# 4. Add vertical lines for Mean and Standard Deviations
plt.axvline(mean, color='red', linestyle='--', label='Mean (0)')
plt.axvline(mean + std_dev, color='green', linestyle='--', label='1 Std Dev')
plt.axvline(mean - std_dev, color='green', linestyle='--')
plt.axvline(mean + 3*std_dev, color='orange', linestyle='--', label='3 Std Dev')
plt.axvline(mean - 3*std_dev, color='orange', linestyle='--')

plt.title('Normal Distribution with 1 and 3 Markers')
plt.legend()
plt.show()
```

### Red Line placed at the mean value.
### Green Lines placed at mean+std and mean-std values. (this range contains the 68% of data)
### Yellow Lines placed at mean+std+std+std and mean-std-std-std values. (this range contains the 99.7% of data)
