```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# 1. Generate Synthetic Data (Blobs)
# We create 3 distinct groups of data
X, y_true = make_blobs(n_samples=300, centers=3, cluster_std=0.60, random_state=0)

# 2. Train K-Means
# We tell it to look for 3 clusters
kmeans = KMeans(n_clusters=3, n_init=10) # n_init=10 runs it 10 times to find best start
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

# 3. Visualization
plt.figure(figsize=(10, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

# Plot the centers (centroids) as black stars
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, marker='*', label='Centroids')

plt.title("K-Means Clustering (K=3)")
plt.legend()
plt.show()
```

### ***K-Means is sensitive to the choice of K. It forces data into spherical blobs.***
